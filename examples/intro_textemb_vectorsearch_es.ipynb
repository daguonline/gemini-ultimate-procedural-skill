# ¿Qué son los Embeddings?

Con el auge de los LLMs, ¿por qué es importante que los ingenieros de TI entiendan cómo funcionan?

En los sistemas de TI tradicionales, la mayoría de los datos se organizan como datos estructurados o tabulares, utilizando palabras clave simples, etiquetas y categorías en bases de datos y motores de búsqueda.

En contraste, los servicios impulsados por IA organizan los datos en una estructura de datos simple conocida como "embeddings".

Una vez entrenada con contenido específico como texto o imágenes, la IA crea un espacio llamado "espacio de embedding", que es esencialmente un mapa del significado del contenido.

La IA puede identificar la ubicación de cada contenido en el mapa; eso es lo que es un embedding.

La IA puede colocar contenidos con significados similares cerca unos de otros en este espacio.

## Casos de uso empresarial de embeddings de texto con LLM

Con la API de embeddings, puedes aplicar la innovación de los embeddings, combinada con la capacidad de los LLM, a diversas tareas de procesamiento de texto, como:

**Búsqueda Semántica habilitada para LLM**: los embeddings de texto se pueden usar para representar tanto el significado como la intención de la consulta de un usuario y los documentos en el espacio de embedding. Los documentos que tienen un significado similar a la intención de búsqueda se encontrarán rápido con tecnología de búsqueda vectorial.

**Clasificación de Texto habilitada para LLM**: se pueden usar para clasificación de texto con una comprensión profunda de diferentes contextos sin necesidad de entrenamiento o ajuste fino (zero-shot learning).

**Recomendación habilitada para LLM**: el embedding de texto puede usarse para sistemas de recomendación como una característica fuerte para entrenar modelos de recomendación.

## Primeros pasos con Vertex AI Embeddings para Texto

### Preparación de Datos

Usaremos el [conjunto de datos público de Stack Overflow](https://console.cloud.google.com/marketplace/product/stack-exchange/stack-overflow) alojado en la tabla de BigQuery `bigquery-public-data.stackoverflow.posts_questions`. Limitaremos el conjunto de datos a 1000 filas para este tutorial.

```python
from google.cloud import bigquery

QUESTIONS_SIZE = 1000

bq_client = bigquery.Client(project=PROJECT_ID)
QUERY_TEMPLATE = """
        SELECT distinct q.id, q.title
        FROM (SELECT * FROM `bigquery-public-data.stackoverflow.posts_questions`
        where Score > 0 ORDER BY View_Count desc) AS q
        LIMIT {limit} ;
        """
query = QUERY_TEMPLATE.format(limit=QUESTIONS_SIZE)
query_job = bq_client.query(query)
rows = query_job.result()
df = rows.to_dataframe()

df.head()
```

### Llamar a la API para generar embeddings

Usaremos la columna `title` y generaremos el embedding para ella.

```python
from google import genai
from google.genai import types

EMBEDDING_MODEL = "gemini-embedding-001"
client = genai.Client(vertexai=True, project=PROJECT_ID, location=LOCATION)
```

En este tutorial usaremos el modelo `gemini-embedding-001`. Puedes pasar hasta 5 textos a la vez en una llamada.

```python
import tqdm
import time
BATCH_SIZE = 5

def get_embeddings_wrapper(texts: list[str]) -> list[list[float]]:
    embeddings: list[list[float]] = []
    for i in tqdm.tqdm(range(0, len(texts), BATCH_SIZE)):
        time.sleep(1)  # evitar error de cuota
        response = client.models.embed_content(
            model=EMBEDDING_MODEL, contents=texts[i : i + BATCH_SIZE]
        )
        embeddings = embeddings + [e.values for e in response.embeddings]
    return embeddings
```

```python
df = df.assign(embedding=get_embeddings_wrapper(list(df.title)))
df.head()
```

## Ver similitudes de embeddings

Veamos cómo se organizan estos embeddings calculando las similitudes de forma rápida. Para el modelo `gemini-embedding-001`, necesitamos usar el producto escalar (dot product).

```python
import random
import numpy as np

key = random.randint(0, len(df))
embs = np.array(df.embedding.to_list())
similarities = np.dot(embs[key], embs.T)
similarities[:5]
```

```python
print(f"Pregunta clave: {df.title[key]}\n")
sorted_questions = sorted(
    zip(df.title, similarities), key=lambda x: x[1], reverse=True
)[:20]
for i, (question, similarity) in enumerate(sorted_questions):
    print(f"{similarity:.4f} {question}")
```
